```python
# code to describe and explore datasets
# this for the houses data
"""['Id', 'MSSubClass', 'MSZoning', 'LotFrontage', 'LotArea', 'Street', 'Alley', 'LotShape', 'LandContour', 
'Utilities', 'LotConfig', 'LandSlope', 'Neighborhood', 'Condition1', 'Condition2', 'BldgType', 'HouseStyle', 
'OverallQual', 'OverallCond', 'YearBuilt', 'YearRemodAdd', 'RoofStyle', 'RoofMatl', 'Exterior1st', 'Exterior2nd', 
'MasVnrType', 'MasVnrArea', 'ExterQual', 'ExterCond', 'Foundation', 'BsmtQual', 'BsmtCond', 'BsmtExposure', 
'BsmtFinType1', 'BsmtFinSF1', 'BsmtFinType2', 'BsmtFinSF2', 'BsmtUnfSF', 'TotalBsmtSF', 'Heating', 'HeatingQC', 
'CentralAir', 'Electrical', '1stFlrSF', '2ndFlrSF', 'LowQualFinSF', 'GrLivArea', 'BsmtFullBath', 'BsmtHalfBath', 
'FullBath', 'HalfBath', 'BedroomAbvGr', 'KitchenAbvGr', 'KitchenQual', 'TotRmsAbvGrd', 'Functional', 'Fireplaces',
'FireplaceQu', 'GarageType', 'GarageYrBlt', 'GarageFinish', 'GarageCars', 'GarageArea', 'GarageQual', 'GarageCond',
'PavedDrive', 'WoodDeckSF', 'OpenPorchSF', 'EnclosedPorch', '3SsnPorch', 'ScreenPorch', 'PoolArea', 'PoolQC', 'Fence',
'MiscFeature', 'MiscVal', 'MoSold', 'YrSold', 'SaleType', 'SaleCondition', 'SalePrice']
"""
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from pandas.api.types import infer_dtype
import os
from math import log 
import seaborn as sns
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
from sklearn.preprocessing import StandardScaler
from sklearn import linear_model
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA
from sklearn.preprocessing import Imputer
import matplotlib.pyplot as plt
%matplotlib inline

def get_n_components(xds,yds):
    # Code source: Gal Varoquaux
    # Modified for documentation by Jaques Grobler
    # License: BSD 3 clause
    # from documentation for sklearn


    logistic = linear_model.LogisticRegression()
    
    pca = PCA()
    pipe = Pipeline(steps=[('pca', pca), ('logistic', logistic)])

    X_digits = xds
    y_digits = yds 

    # Plot the PCA spectrum
    pca.fit(X_digits)

    plt.figure(1, figsize=(4, 3))
    plt.clf()
    plt.axes([.2, .2, .7, .7])
    plt.plot(pca.explained_variance_, linewidth=2)
    plt.axis('tight')
    plt.xlabel('n_components')
    plt.ylabel('explained_variance_')

    # Prediction
    n_components = [3, 6, 9]
    Cs = np.logspace(-4, 4, 3)

    # Parameters of pipelines can be set using __ separated parameter names:
    estimator = GridSearchCV(pipe,
                             dict(pca__n_components=n_components,
                                  logistic__C=Cs))
    estimator.fit(X_digits, y_digits)

    plt.axvline(estimator.best_estimator_.named_steps['pca'].n_components,
                linestyle=':', label='n_components chosen')
    plt.legend(prop=dict(size=12))
    plt.show()
    print('Suggest use %d components' % estimator.best_estimator_.named_steps['pca'].n_components)



def doheat(ds,fontsize,saveas):
    # simple heatmap of pairwise correlations for all (numeric) variables
    # https://stackoverflow.com/questions/32244753/how-to-save-a-seaborn-plot-into-a-file/39482402#39482402
    # cm = plt.cm.PiYG # plt.cm.RdBu  plt.cm.Blues
    dsc = ds.corr(method='spearman')
    plt.subplots(figsize=(36,36))
    hmap = sns.heatmap(dsc, square=True, cmap=plt.cm.RdBu,xticklabels="auto",yticklabels="auto")
    fig = hmap.get_figure()
    #ax.set_xlabel('text', fontdict={'fontsize' : fontsize})
    plt.show()
    fig.savefig(saveas)


def dohists(ds):
    num_bins = 50
    dtyps = ds.dtypes.values
    for i,c in enumerate(list(ds.columns.values)):
        x = ds[c].values
        t = dtyps[i]
        print('###%d' % i,c,'is',t)
        ctab = ds[c].value_counts()
        if (ctab.size > 2) and (str(t) != 'category'):
            num_bins=min(50,ctab.size+1)
            fig, ax = plt.subplots()
            n, bins, patches = ax.hist(x, density=1)
            #n, bins, patches = ax.hist(x[~np.isnan(x)], density=1)
            ax.set_xlabel(c)
            ax.set_ylabel('Probability density')
            ax.set_title(r'Histogram of %s' % c)

            # Tweak spacing to prevent clipping of ylabel
            fig.tight_layout()
            ax.grid()
            plt.show()
            #fig.savefig("test.png")
        else:
            print('%s has distribution: %s' % (c,ctab))


def npca3d(data,labs,mytitle):
    fig = plt.figure(figsize=(8, 6))
    ax = Axes3D(fig, elev=-150, azim=140)
    #ax = fig.add_subplot(111, projection='3d')
    pca = PCA(n_components=3)
    
    df = pca.fit_transform(data)
    d = pd.DataFrame({"x":df[:, 0],"y":df[:,1],"z":df[:,2],"pq":labs})

    labels = ['q1','q2','q3','q4','q5']
    cs = ['red','orange','yellow','green','blue']

    for i, dff in d.groupby("pq"):
        plt.scatter(dff["x"], dff["y"], dff["z"], c=cs[i], #cmap=plt.cm.PiYG, 
                    edgecolors='k', label="Q%d" % i)

    ax.set(xlabel="1st eigenvector", ylabel="2nd eigenvector",zlabel="3rd eigenvector",
           title=mytitle)
    ax.grid()
    ax.legend()
    #fig.savefig("test.png")
    plt.show()

def pca3d(data,labs,mytitle):

    pca = PCA(n_components=3)
    labels = ['q1','q2','q3','q4','q5']
    cs = ['red','orange','yellow','green','blue']
    colrs = [cs[x] for x in labs]

    #cmap = ['red','orange','blue','lightgray']

    fig = plt.figure(figsize=(8, 6))
    ax = Axes3D(fig, elev=-150, azim=120)

    df = pca.fit_transform(data)
    d = pd.DataFrame({"x":df[:, 0],"y":df[:,1],"z":df[:,2],"pq":labs})

    labels = ['q1','q2','q3','q4','q5']
    cs = ['red','orange','yellow','green','blue']
    for i, dff in d.groupby("pq"):
        plt.scatter(dff['x'], dff["y"],   dff["z"],  c=cs[i], cmap=plt.cm.PiYG, 
                    edgecolors='k', label="Q%d" % i)

    ax.set(xlabel="1st eigenvector", ylabel="2nd eigenvector",zlabel="3rd eigenvector",
           title='Samples in first 3 eigenvector space')
    #X = ax.plot(d['x'], d["y"],   d["z"],  'ko') # X is a *list* of line2d objects...

    #X[0].set_markersize(5) # set_ms works too 
    ax.grid()
    ax.legend()

    #fig.savefig("test.png")
    plt.show()
    
def pretty_print_linear(coefs, names, sort=False):
    
    lst = zip(coefs, names)
    if sort:
        lst = sorted(lst, key=lambda x: -np.abs(x[0]))
    return " + ".join("%s * %s" % (round(coef, 5), name)
                      for coef, name in lst if (coef != 0.0))


def pca2d(data,labs,mytitle):

    pca = PCA(n_components=2)  
    vnames = all_data.columns
    df = pca.fit_transform(data)
    d = pd.DataFrame({"x":df[:, 0],"y":df[:,1],"pq":labs})

    labels = ['q1','q2','q3','q4','q5']
    cs = ['red','orange','yellow','green','blue']
    fig, ax = plt.subplots()

    for i, dff in d.groupby("pq"):
        plt.scatter(dff['x'], dff["y"], s=20, c=cs[i], cmap=plt.cm.PiYG,
                    edgecolors='k', label="Q%d" % i)

    ax.set(xlabel="1st eigenvector", ylabel="2nd eigenvector",
           title=mytitle)
    ax.grid()
    ax.legend()
    #fig.savefig("test.png")
    plt.show()
        
    nx = data.shape[1]
    nx = min(10,nx)
    xvals = [x+1 for x in range(0,nx)]
    X = pd.DataFrame({'x':xvals})
    X = X.iloc[:,0]
    tix = xvals
    pca = PCA(n_components=nx)
    pca.fit(data)
    
    ax = sns.scatterplot(x=X,y=pca.explained_variance_ratio_)
    ax.set(xlabel="PCA component",ylabel="Proportion of variance explained")
    
    plt.xticks(rotation=90)
    plt.show()

            
def data_cleaning_housing(data):
    """ from a kernel. good enough"""
    data = data.drop(data[(data['GrLivArea']>4000) & (data['SalePrice']<300000)].index)
    for col in ['PoolQC',"MiscFeature",'Alley','Fence',"FireplaceQu",'GarageType', 
           'GarageFinish', 'GarageQual', 'GarageCond','BsmtQual', 'BsmtCond', 
            'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2']:
        data[col] = data[col].fillna('None')
    for col in ('GarageYrBlt', 'GarageArea', 'GarageCars','BsmtFinSF1', 'BsmtFinSF2','LotFrontage',
            'BsmtUnfSF','TotalBsmtSF', 'BsmtFullBath', 'BsmtHalfBath',"MasVnrArea"):
        data[col] = data[col].fillna(0)
    
    return data

data=pd.read_csv('train.csv')
data = data.drop(['Id'],axis=1)
#ddata.replace([np.inf, -np.inf], np.nan)
ddata = data.replace([np.inf, -np.inf], np.nan).dropna(how="all")
ddata = data_cleaning_housing(ddata)
target = ddata[['SalePrice']].iloc[:,0]
ltarget = np.log(target)
#for i,p in enumerate(target):
#    ltarget[i] = log(p) # no zeros and skewed so let's predict this as a continuous variable

#mssc = ddata['MSSubClass'].value_counts()
#for (labl,valu) in [('msc20',20),('msc60',60),('msc50',50),('msc120',120)]:
#    ddata[labl] = (ddata['MSSubClass'] == valu) #[20,60,50,120] are the commonest
#ddata = ddata.drop(['Id','MSSubClass'], axis=1)
tndata = ddata.select_dtypes(exclude='object')
nonnumdata = ddata.select_dtypes(include='object')
for colm in nonnumdata.columns:
    ddata[colm] = ddata[colm].astype('category')

    

pricequintile = pd.qcut(ddata['SalePrice'], 5, labels=False) #['saleq1','saleq2','saleq3','saleq4','saleq5'])

#ndata = tndata.drop(['SalePrice','msc120','msc20','msc50','msc60'], axis=1)
ndata = tndata.drop(['SalePrice'], axis=1)

#print(list(ddata.columns.values))
#clean = ndata[~np.isnan(ndata)]
print(ddata.shape)
all_data = pd.get_dummies(ddata)
print('After get_dummies',all_data.shape)
sall_data = StandardScaler().fit_transform(all_data)
#get_n_components(sall_data,pricequintile)

pca2d(sall_data,pricequintile,'Housing 2d PCA')
pca3d(sall_data,pricequintile,'Housing 3d PCA')

doheat(tndata,10,'house_heat.png')

doheat(all_data,4,'houseall_heat.png')

dohists(ddata)


```

```python
from sklearn.linear_model import LassoCV
from sklearn.linear_model import Lasso
from sklearn.model_selection import KFold
from sklearn.model_selection import GridSearchCV


# A helper method for pretty-printing linear models
def pretty_print_linear(coefs, names, sort=False):
    
    lst = zip(coefs, names)
    if sort:
        lst = sorted(lst, key=lambda x: -np.abs(x[0]))
    return " + ".join("%s * %s" % (round(coef, 5), name)
                      for coef, name in lst if (coef != 0.0))

rtrain = all_data.drop(columns = ['Id'])
y_train = np.log(rtrain['SalePrice'])
rtrain = rtrain.drop(columns = ['SalePrice'])
nms = rtrain.columns
train = StandardScaler().fit_transform(rtrain)
#train = rtrain

lasso = Lasso(random_state=None)
alphas = np.logspace(-3, -0.5, 20)

tuned_parameters = [{'alpha': alphas}]
n_folds = 3

clf = GridSearchCV(lasso, tuned_parameters, scoring="neg_mean_squared_error", cv=n_folds, refit=False)
clf.fit(train, y_train)
scores = clf.cv_results_['mean_test_score']
scores_std = clf.cv_results_['std_test_score']
plt.figure().set_size_inches(8, 6)
plt.semilogx(alphas, scores)

# plot error lines showing +/- std. errors of the scores
std_error = scores_std / np.sqrt(n_folds)

plt.semilogx(alphas, scores + std_error, 'b--')
plt.semilogx(alphas, scores - std_error, 'b--')

# alpha=0.2 controls the translucency of the fill color
plt.fill_between(alphas, scores + std_error, scores - std_error, alpha=0.2)

plt.ylabel('CV score +/- std error')
plt.xlabel('alpha')
plt.axhline(np.max(scores), linestyle='--', color='.5')
plt.xlim([alphas[0], alphas[-1]])

plt.show()

# #############################################################################
# Bonus: how much can you trust the selection of alpha?

# To answer this question we use the LassoCV object that sets its alpha
# parameter automatically from the data by internal cross-validation (i.e. it
# performs cross-validation on the training data it receives).
# We use external cross-validation to see how much the automatically obtained
# alphas differ across different cross-validation folds.
#lasso_cv = LassoCV(alphas=alphas, fit_intercept=False, random_state=None)
lasso_cv = LassoCV( fit_intercept=False, random_state=None)
k_fold = KFold(n_splits=3,shuffle=True)

print("Answer to the bonus question:",
      "how much can you trust the selection of alpha?")
print()
print("Alpha parameters maximising the generalization score on different")
print("subsets of the data:")
for k,(ftrain, ftest) in enumerate(k_fold.split(train)):
    xtr = train[ftrain]
    ytr = y_train[[ftrain]]
    print('ytr is',ytr)
    lasso_cv.fit(xtr, ytr)
    xt = train[ftest]
    yt = y_train[[ftest]]
    lt = lasso_cv.score(xt,yt)
    print("[fold {0}] alpha: {1:.5f}, score: {2:.5f}".
          format(k, lasso_cv.alpha_, lt))
print()
print("Answer: Not very much since we obtained different alphas for different")
print("subsets of the data and moreover, the scores for these alphas differ")
print("quite substantially.")
```
